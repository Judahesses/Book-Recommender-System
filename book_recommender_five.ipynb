> https://www.goodreads.com/api

```python
from xmljson import badgerfish as bf
from xml.etree.ElementTree import fromstring
from json import dumps
import json
```

```python
def get_data(user_count):
    dataset = {}
    complete_set = []
    for i in range(20000, user_count+1):
        url = 'https://www.goodreads.com/review/list?v=2'
        key = api
        url_params = {
                        'key': key,
                        'id': i,
                        'per_page': 200
                    }
        response = requests.get(url, params=url_params)
        dataset[i] = json.loads(dumps(bf.data(fromstring(response.content))))

        try:
            for j in range(len(dataset[i]['GoodreadsResponse']['reviews']['review'])):
                    if dataset[i]['GoodreadsResponse']['reviews']['review'][j]['rating']['$'] > 0:
                        complete_set.append(dict([
                        ('User',i),
                        ('Book', dataset[i]['GoodreadsResponse']['reviews']['review'][j]['book']['title']['$']), 
                        ('Rating', dataset[i]['GoodreadsResponse']['reviews']['review'][j]['rating']['$'])]))
        except:
            pass
        time.sleep(1)
        
        #see where we're at
        if i % 50 == 0:
            print(i)

        if response.status_code != 200:
            print(i, response.status_code)         
            
    df_five = pd.DataFrame(complete_set)
    df_five.to_csv('fifth_book_data')
    df_five = pd.read_csv('fifth_book_data')
    return df_five
```

```python
start = time.time()
df_five = get_data(25000)
end = time.time()
print(end - start)
```
